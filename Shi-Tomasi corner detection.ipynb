```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\Users\Aravind\Downloads\cv course\DATA\flat_chessboard.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray=cv2.imread(r'C:\Users\Aravind\Downloads\cv course\DATA\flat_chessboard.png',0)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
img_1=cv2.imread(r'C:\Users\Aravind\Downloads\cv course\DATA\real_chessboard.jpg')
img_1=cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
gray_1=cv2.imread(r'C:\Users\Aravind\Downloads\cv course\DATA\real_chessboard.jpg',0)
```

```python
plt.imshow(gray_1,cmap='gray')
```

```python
corners=cv2.goodFeaturesToTrack(gray,5,0.01,10)
```

```python
corners=np.int0(corners)
```

```python
corners
```

```python
for i in corners:
    x,y=i.ravel()
    cv2.circle(img,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(img)
```
